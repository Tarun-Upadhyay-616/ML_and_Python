```python
import json
```

```python
def load_data(filename):
    with open(filename,'r') as f:
        data = json.load(f)
    return data
```

```python
data = load_data("store_data.json")
print(data)
```

```python
def Data_Cleanup(data):
    reviews_to_num = {"zero": 0,"one":1,"two":2,"three":3,"four":4,"five":5}
    for user in data:
        # rating clear
        user['review'] = (str(user['review']))
        raw_review = user['review'].strip().lower()
        # print(raw_review)
        if(raw_review in reviews_to_num):
            user['review'] = reviews_to_num[raw_review]
        print(user['review'])
Data_Cleanup(data)
```

```python
def age_correction(data):
    for user in data:
        print(type(user['age']))
        user['age'] = int(user['age'])
        if(user['age']<0 or user['age'] == None):
            user['age'] = 0
        
age_correction(data)
```
